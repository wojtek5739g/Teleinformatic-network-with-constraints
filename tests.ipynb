<h1><span style="color:blue">First Algorithm</span></h1>

1.  Roulette preselection
2.  Uniform mutation
3.  Two-points crossover
4.  Succession with partial replacement

```python
mut_rates = [0.001, 0.5] 
cross_rates = [0.5, 0.75]
num_of_iterations_list = [100, 800]
pop_sizes = [150, 250]
modularities = [5]
dissagregations = [False]


# Iteration through all combinations
for mut_rate, cross_rate, num_of_iterations, pop_size, modularity, dissagregation in itertools.product(mut_rates, cross_rates, num_of_iterations_list, pop_sizes, modularities, dissagregations):
    systems_end_max = start_evolutionary_algorithm(1, 1, dissagregation, pop_size, nodes, links, demands_paths, demands_values, num_of_iterations, 2, mut_rate, cross_rate, 0, modularity, True)
    
    frame_one.loc[len(frame_one)] = [mut_rate, cross_rate, num_of_iterations, pop_size, modularity, dissagregation, systems_end_max]
```
